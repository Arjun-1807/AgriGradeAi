```python
from ultralytics import YOLO
import kagglehub
import os
from pathlib import Path
```

```python
dataset_path = kagglehub.dataset_download("moltean/fruits")

print("Dataset downloaded to:", dataset_path)
```

```python
dataset_root = Path(dataset_path) / "fruits-360_100x100" / "fruits-360"

print("Dataset root:", dataset_root)

print("Folders inside dataset:")
print(os.listdir(dataset_root))
```

```python
train_dir = dataset_root / "Training"

print("Training directory:", train_dir)
```

```python
classes = sorted(os.listdir(train_dir))

print("Total classes:", len(classes))
```

```python
for cls in classes:
    print(cls)
```

```python
for cls in classes:

    class_path = train_dir / cls
    num_images = len(list(class_path.glob("*")))

    print(f"{cls}: {num_images} images")
```

```python
import shutil
from pathlib import Path

selected_classes = [
    "Apple Granny Smith 1",
    "Banana 1",
    "Cherry 1",
    "Grape Blue 1",
    "Kiwi 1",
    "Lemon 1",
    "Mango 1",
    "Orange 1",
    "Peach 1",
    "Pear 1",
    "Pineapple 1",
    "Pomegranate 1",
    "Strawberry 1",
    "Watermelon 1",
    "Carrot 1",
    "Cucumber 1",
    "Eggplant 1",
    "Onion Red 1",
    "Pepper Red 1",
    "Tomato 1"
]

target_dataset = Path("/content/produce_dataset")

for cls in selected_classes:

    src = train_dir / cls
    dst = target_dataset / cls

    dst.mkdir(parents=True, exist_ok=True)

    for img in list(src.glob("*"))[:600]:
        shutil.copy(img, dst)

print("Dataset ready!")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n-cls.pt")

model.train(
    data="/content/produce_dataset",
    epochs=20,
    patience=5,
    imgsz=224,
    batch=32,
    device=0,
    workers=2
)
```
