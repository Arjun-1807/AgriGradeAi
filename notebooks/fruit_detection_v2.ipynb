```python
import yaml

yaml_path = "data.yaml"

with open(yaml_path, "r") as f:
    data = yaml.safe_load(f)

# Fix the paths
data["train"] = "train/images"
data["val"] = "valid/images"

# Save back
with open(yaml_path, "w") as f:
    yaml.dump(data, f)

print("YAML file updated successfully!")
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="data.yaml",
    epochs=80,
    imgsz=640,
    batch=32,
    cache=True,
    workers=8,
    patience=20,
    device=0
)
```
